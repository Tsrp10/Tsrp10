```python
print("Star pattern")
for i in range(5):
  for j in range(5):
    print("*", end=" ")
  print()
```

```python
for i in range(5):
  for j in range(i):
    print("*",end=" ")
  print()
```

```python
for i in range(5,0,-1):
  for j in range(i):
    print("*",end=" ")
  print()
```

```python
for i in range(1,6):
  print(i,end=" ")
```

```python
for i in range(6):
  for j in range(1,i+1):
    print(j,end=" ")
  print()
```

```python
for i in range(5,0,-1):
  for j in range(1,i+1):
    print(j,end=" ")
  print()
```

```python
num=1
for i in range(5):
  for j in range(i):
    print(num,end=" ")
    num +=1
  print()
```
